## Introduzione a python e calcolare  π

Per trovare il valore di π possiamo simulare una partita di freccette, in cui su un numero n di lanci, solo un numero x arriva sul bersaglio a forma di cerchio. Se supponiamo che questo cerchio abbia r = 1, sappiamo che la sua area è π. Se anzichè considerare l'intera superficie del cerchio, consideriamo solo il quarto inscritto in un quadrato di l = 1, sappiamo che l'area di quella parte di cerchio sarà π/4. L'area di quel quarto di cerchio, sarà uguale al rapporto fra il numero di freccette nel cerchio e il numero totale di freccette. Moltiplicando il tutto per 4 otterremo il valore di π, secondo una precisione che varierà in un modo che spiegheremo più avanti.
Per capire meglio la situazione, osserviamo il disegno sotto riportato.


```python
import matplotlib.pyplot as plt
import numpy.random as rnd
from matplotlib.patches import Ellipse

%matplotlib inline


ells = Ellipse(xy=(0,0), width=2, height=2, angle=0.0)


fig = plt.figure(0)
ax = fig.add_subplot(111, aspect='equal')

ax.add_artist(ells)
ells.set_clip_box(ax.bbox)
ells.set_facecolor("red")
ells.set_alpha(0.3)

plt.axvline(0, color="black")
plt.axhline(0, color="black")

sqr = plt.Rectangle(xy=(0,0), width=1, height=1, angle=0.0)

fig = plt.figure(0)
ax = fig.add_subplot(111, aspect='equal')

ax.add_artist(sqr)
sqr.set_alpha(0.3)


ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

plt.show()
```

Come possiamo notare, la funzione *random()* ci restituisce un valore casuale.

Ora, creiamo una *lista* (che viene identificata dalle parentesi quadre) che contenga una serie di coppie di numeri casuali che rappresenteranno le nostre freccette


```python
darts_x = [(random()) for x in range(1000)]
```


In questo modo abbiamo chiesto a python di generarci una lista composta
da 1000 elementi che saranno le coordinate x delle nostre freccette
e sono numeri casuali

```python
darts_x[:10]
```


Verifichiamo che i valori siano effettivamente 1000 utilizzando la funzione *len()* che ci restituisce la lunghezza della lista "darts"


```python
len(darts_x)
```

Facciamo ora la stessa cosa per ottenere le y dei punti


```python
darts_y = [(random()) for x in range(1000)]
```

```python
darts_y[:10]
```

Ora che abbiamo le nostre coordinate x e y, possiamo rappresentarle
graficamente usando la funzione "matplotlib.pylot"


```python
import matplotlib.pyplot as plt
```

```python

%matplotlib inline
_ = plt.plot([darts_x],[darts_y], "ro")

```

In questo modo abbiamo ottenuto il disegno della simulazione della nostra partita.

Creiamo ora una lista che abbia come elementi delle coppie di coordinate, in cui le ascisse sono progressivamente gli elementi di "darts_x" e le ordinate gli elementi di "darts_y": come fare? basta dire che nella lista deve esserci una coppia in cui il primo elemento appartiene a "darts_x" e il secondo a "darts_y" e che deve fare questo per ogni elemento delle due liste, quindi a partire dal primo[i] in tutto l'intervallo di lunghezza "darts_x"



```python
darts_xy = [(darts_x[i], darts_y[i]) for i in range(len(darts_x))]
```

Chiamiamo la lista per vedere ciò che abbiamo ottenuto

```python
darts_xy[:10]
```

Vogliamo ora vedere quali freccette sono effettivamente entrate nel bersaglio: sappiamo che l'equazione della circonferenza di r = 1 è

*x^2 + y^2 = 1*

Quindi, le freccette nel cerchio saranno quelle che avranno *x^2 + y^2 < 1*


```python
darts_in_circle = [d for d in darts_xy if d[0]**2+d[1]**2<1]
```

Chiamiamo "darts_in_circle" per vedere quali siano queste freccette

```python
darts_in_circle[:10]
```

Se vogliamo unire i due grafici in modo da visualizzare graficamente quali siano le freccette nella parte di cerchio che stiamo considerando e quali invece non lo siano, dobbiamo "dividere" la lista "darts_in_circle" in modo da avere il primo elemento della prima coppia come ascissa e il secondo della stessa come ordinata

```python
darts_in_circle_x = [x[0] for x in darts_in_circle]
```

Chiamiamo "darts_in_circle_x" per verificare di aver ottenuto ciò che desideravamo

```python
darts_in_circle_x[:10]
```

```python
darts_in_circle_y = [x[1] for x in darts_in_circle]
```

E ora disegniamo il grafico

```python
%matplotlib inline
_ = plt.plot([darts_x],[darts_y], "bo")
_ = plt.plot(darts_in_circle_x, darts_in_circle_y, "ro")
```

Chiediamo ora a python di contare gli elementi in "darts_in_circle", cioè il numero di freccette nel quarto di cerchio che stiamo considerando

```python
len(darts_in_circle)
```
