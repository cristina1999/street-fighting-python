```python
from random import random
```

```python
# in questo modo importiamo la funzione "random" ("import random")
# dalla libreria "random" ("from random") che si aggiungerà alle funzioni
# base già presenti in python

# ma cosa fa la funzione "random"? Per scoprirlo, basta chiamarla
```

```python
random()
```

```python
# come possiamo notare, la funzione random ci restituisce un valore casuale

# ora creiamo una lista (che viene identificata dalle parentesi quadre)
# che contenga una serie di coppie di numeri casuali
```

```python
darts_x = [(random()) for x in range(1000)]
```

```python
# in questo modo abbiamo chiesto a python di generarci una lista composta
# da 1000 elementi che saranno le coordinate x dei nostri punti
# e sono numeri casuali
```

```python
darts_x
```

```python
# verifichiamo che i valori siano effettivamente 1000 utilizzando
# la funzione "len(darts_x)" che ci restituisce la
# lunghezza della lista "darts_x"
```

```python
len(darts_x)
```

```python
# facciamo ora la stessa cosa per ottenere le y dei punti
```

```python
darts_y = [(random()) for x in range(1000)]
```

```python
darts_y
```

```python
# ora che abbiamo le nostre coordinate x e y, possiamo rappresentarle
# graficamente usando la funzione "matplotlib.pylot"
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
_ = plt.plot([darts_x],[darts_y], "ro")
```

```python
darts_xy = [(darts_x[i], darts_y[i]) for i in range(len(darts_x)) ]
```

```python
darts_xy
```

```python
darts_in_circle = [d for d in darts_xy if d[0]**2+d[1]**2<1]
```

```python
darts_in_circle
```

```python
darts_in_circle_x = [x[0] for x in darts_in_circle]
```

```python
darts_in_circle_x
```

```python
darts_in_circle_y = [x[1] for x in darts_in_circle]
```

```python
%matplotlib inline
_ = plt.plot([darts_x],[darts_y], "bo")
_ = plt.plot(darts_in_circle_x, darts_in_circle_y, "ro")
```

```python
len(darts_in_circle)
```
